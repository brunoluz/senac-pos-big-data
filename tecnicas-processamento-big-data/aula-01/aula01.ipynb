```python
import os
import gdown
import pandas as pd
```

```python
url_enale = 'https://drive.google.com/uc?id=1o1LzQySzgWrDcSv-TxiiegehgS2axFdb'
file_microdados_enade_2019 = 'microdados_enade_2019.txt'

gdown.download(url_enale, file_microdados_enade_2019, quiet=False)
```

```python
url_alunos = 'https://drive.google.com/uc?id=1mBDuh_cp3dtscg3pMsbZNjA2_bsT3Mu4'
file_microdados_alunos_2019 = 'microdados_enem_2019_alunos_SP.csv'

gdown.download(url_alunos, file_microdados_alunos_2019, quiet=False)
```

```python
df_enade_2019 = pd.read_csv(file_microdados_enade_2019, sep=';', decimal=',')
```

```python
df_enade_2019.head(10)
```

```python
df_alunos_2019 = pd.read_csv(file_microdados_alunos_2019, sep=',', decimal='.')
```

```python
df_alunos_2019.head(10)
```

1. Qual é a média da nota em matemática de todos os alunos de São Paulo?

```python
nota_matematica = df_alunos_2019.loc[
        df_alunos_2019.SG_UF_RESIDENCIA == 'SP'
    ].NU_NOTA_MT.mean()
print(f"A média da nota de matemática de todos os alunos de são paulo é: {nota_matematica:.2f}")
```

2. Qual é a média da nota em Linguagens e Códigos de todos os alunos de SP?

```python
nota_linguagens_codigo = df_alunos_2019.loc[
        df_alunos_2019.SG_UF_RESIDENCIA == 'SP'
    ].NU_NOTA_LC.mean()
print(f"A média da nota de Linguagens e Códigos de todos os alunos de são paulo é: {nota_linguagens_codigo:.2f}")
```

3. Qual é a média da nota em Ciências Humanas dos alunos do sexo FEMININO de SP?

```python
nota_ciencias_humanas_feminino = df_alunos_2019.loc[
        (df_alunos_2019.SG_UF_RESIDENCIA == 'SP') & 
        (df_alunos_2019.TP_SEXO == 'F')
    ].NU_NOTA_CH.mean()
print(f"A média da nota em Ciências Humanas dos alunos do sexo FEMININO de SP é: {nota_ciencias_humanas_feminino:.2f}")
```

4. Qual é a média da nota em Matemática dos alunos do sexo FEMININO que moram na cidade de Santos?

```python
nota_matematica_feminino_santos = df_alunos_2019.loc[
        (df_alunos_2019.SG_UF_RESIDENCIA == 'SP') & 
        (df_alunos_2019.TP_SEXO == 'F') &
        (df_alunos_2019.NO_MUNICIPIO_RESIDENCIA == 'Santos')
    ].NU_NOTA_MT.mean()
print(f"A média da nota em Matemática dos alunos do sexo FEMININO que moram na cidade de Santos é: {nota_matematica_feminino_santos:.2f}")
```

5. Qual é a média da nota em Matemática dos alunos do município de Campinas que possuem TV por assinatura na residência?

```python
nota_matematica_campinas_tv_assinatura = df_alunos_2019.loc[
        (df_alunos_2019.SG_UF_RESIDENCIA == 'SP') & 
        (df_alunos_2019.NO_MUNICIPIO_RESIDENCIA == 'Campinas') & 
        (df_alunos_2019.Q021 == 'B') 
    ].NU_NOTA_MT.mean()
print(f"A média da nota em Matemática dos alunos do município de Campinas que possuem TV por assinatura na residência é: {nota_matematica_campinas_tv_assinatura:.2f}")
```

6. Qual é a média da nota em Ciências Humanas dos alunos que possuem dois fornos micro-ondas em casa?

```python
nota_ciencias_humanas_dois_microondas = df_alunos_2019.loc[
        (df_alunos_2019.Q016 == 'C') 
    ].NU_NOTA_CH.mean()
print(f"A média da nota em Ciências Humanas dos alunos que possuem dois fornos micro-ondas em casa é: {nota_ciencias_humanas_dois_microondas:.2f}")
```

7. Qual é a nota média em Matemática dos alunos de Brotas e de Jaú?

```python
nota_matematica_brotas_jau = df_alunos_2019.loc[
        (df_alunos_2019.NO_MUNICIPIO_RESIDENCIA == 'Brotas') |
        (df_alunos_2019.NO_MUNICIPIO_RESIDENCIA == 'Jaú') 
    ].NU_NOTA_MT.mean()
print(f"A nota média em Matemática dos alunos de Brotas e de Jaú é: {nota_matematica_brotas_jau:.2f}")
```

8. Qual é a nota média em Ciências Humanas dos alunos de SP que moram sozinhos?

```python
nota_ciencias_humanas_moram_sozinho = df_alunos_2019.loc[
        (df_alunos_2019.SG_UF_RESIDENCIA == 'SP') & 
        (df_alunos_2019.Q005 == 1)
    ].NU_NOTA_CH.mean()
print(f"A nota média em Ciências Humanas dos alunos de SP que moram sozinhos é: {nota_ciencias_humanas_moram_sozinho:.2f}")
```

9. Qual é a nota média em Ciências Humanas dos alunos de SP cujo pai completou Pós-graduação e possuem renda familiar entre R$ 8.982,01 e R$ 9.980,00.

```python
nota_ciencias_humanas_sp_pai_graduacao_renda_alta = df_alunos_2019.loc[
        (df_alunos_2019.SG_UF_RESIDENCIA == 'SP') & 
        (df_alunos_2019.Q001 == 'G') &
        (df_alunos_2019.Q006 == 'M')
    ].NU_NOTA_CH.mean()
print(f"A nota média em Ciências Humanas dos alunos de SP cujo pai completou Pós-graduação e possuem renda familiar entre R$ 8.982,01 e R$ 9.980,00 é: {nota_ciencias_humanas_sp_pai_graduacao_renda_alta:.2f}")
```

10. Qual é a nota média em Matemática dos alunos do sexo Feminino que moram em Ourinhos e escolheram “Espanhol” como língua estrangeira?

```python
nota_matematica_feminio_ourinhos_espanhol = df_alunos_2019.loc[
        (df_alunos_2019.TP_SEXO == 'F') & 
        (df_alunos_2019.NO_MUNICIPIO_RESIDENCIA == 'Ourinhos') & 
        (df_alunos_2019.TP_LINGUA == 1)
    ].NU_NOTA_MT.mean()
print(f"A nota média em Matemática dos alunos do sexo Feminino que moram em Ourinhos e escolheram “Espanhol” como língua estrangeira é: {nota_matematica_feminio_ourinhos_espanhol:.2f}")
```
